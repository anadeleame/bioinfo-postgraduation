```python
output_file = open('output_file.txt', 'w')
output_file.write('chrom\tpos\tref\talt\tcol_filter\tfraction\tmq\tmqsum\treadsum\thotspot\tgermline\n')

with open('output_somatico.vcf', 'r') as vcf:
  for line in vcf:
    if not line.startswith('#'):
      linesplit = line.split('\t')
      chr = linesplit[0]
      pos = linesplit[1]
      ref = linesplit[3]
      alt = linesplit[4]
      filt = linesplit[6]
      info = linesplit[7]
      infosplit = info.split(';')

      fraction = False
      mq = False
      mqsum = False                                #starting variables with false so they can always exist
      readsum = False
      hotspot = False
      germline = False

      for data in infosplit:
        data_w_equal = data.split('=')
        userlist = data_w_equal[0]                    #what the user wants is or can be in element 0. We don't have to increment because everytime that the loop occurs, it will pass to the next value and this one will always be the first, having index 0
        if 'FractionInformativeReads' == userlist:
          fraction = data_w_equal[1]                  #we split with '=', its name is in the position 0 and its value in the position 1
        elif 'MQ' == userlist:                        #if the condition above don't be accepted, the loop starts here. But the variants wouldn't be always created if they are not there, and they need to to be in the final output file. That's why I declared them above before 'for'
          mq = data_w_equal[1]
        elif 'MQRankSum' == userlist:
          mqsum = data_w_equal[1]
        elif 'ReadPosRankSum' == userlist:
          readsum = data_w_equal[1]
        elif 'hotspot' == userlist:
          hotspot = True
        elif 'GermlineStatus' == userlist:
          germline = data_w_equal[1]
      output_file.write(f'{chr}\t{pos}\t{ref}\t{alt}\t{filt}\t{fraction}\t{mq}\t{mqsum}\t{readsum}\t{hotspot}\t{germline}\n')

output_file.close()
```
